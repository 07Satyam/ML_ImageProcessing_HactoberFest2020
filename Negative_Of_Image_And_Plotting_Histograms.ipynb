```python
for i in range(0, im.size[0] - 1): 
    for j in range(0, im.size[1] - 1): 
        pixel = im.getpixel((i,j))
        pixel1 = 255 - pixel[0] 
        pixel2 = 255 - pixel[1] 
        pixel3 = 255 - pixel[2] 
        im.putpixel( (i,j),(pixel1,pixel2,pixel3) )
  
# Display the negative transformed image 
plt.imshow(im) 
plt.show() 
```

```python
from PIL import Image
import numpy as np
import matplotlib.pylab as plt
im_g = im.convert('L')
plt.imshow(im_g)
```

```python
im_g=np.array(im_g)
```

```python
c = 255/(np.log(1 + np.max(im_g))) 
log_transformed = c * np.log(1 + im_g) 
  
log_transformed = np.array(log_transformed, dtype = np.uint8) 
  
plt.title("log transformation")
plt.imshow(log_transformed, cmap=plt.cm.gray)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("Images1/parrot.png")
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.exposure import equalize_hist
img = cv2.imread("Images1/parrot.png")
plt.imshow(img)
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
eq = np.float32(equalize_hist(img))
plt.imshow(eq)
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([eq],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
from PIL import Image, ImageEnhance
img = Image.open("Images1/parrot.png")
new=ImageEnhance.Contrast(img)
res=new.enhance(2)
plt.imshow(res)
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([np.float32(res)],[i],None,[255],[0,255])
    plt.plot(histr,color = col)
    plt.xlim([0,255])
plt.show()
```
